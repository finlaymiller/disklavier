```python
import os
from pretty_midi import PrettyMIDI, Instrument, Note
import numpy as np
import pandas as pd
```

```python
dataset_path = os.path.join("..", "data", "datasets", "careful")
output_path = os.path.join("..", "data", "datasets", "careful_transpositions.parquet")
transformations = list(range(12))
```

```python
def transpose_midi(input_file_path: str, semitones: int) -> np.ndarray:
    midi = PrettyMIDI(input_file_path)
    for instrument in midi.instruments:
        for note in instrument.notes:
            note.pitch += semitones

    return midi.get_pitch_class_histogram()
```

```python
filenames = [f for f in os.listdir(dataset_path) if f[-4:] == ".mid"]

df = pd.DataFrame(index=filenames, columns=transformations)

for segment in filenames:
    print(f"transforming {segment}")
    segment_path = os.path.join(dataset_path, segment)

    for t in transformations:
        df.at[segment, t] = transpose_midi(segment_path, t)
```

```python
df.head()
df.to_parquet(output_path)
```
