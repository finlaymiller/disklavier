```python
import os
import faiss
import numpy as np
from glob import glob
```

```python
faiss_index = faiss.read_index(
    "/Users/finlay/Documents/Programming/disklavier/data/tables/20250320/clap-sgm.faiss"
)
input_path = (
    "/Users/finlay/Documents/Programming/disklavier/data/datasets/20250320/segmented"
)
filenames = list(glob(os.path.join(input_path, "*.mid")))
filenames.sort()
print(f"found {len(filenames)} files:\n\t{filenames[:5]}")
```

```python
# ensure that embedding is normalized
query_embedding = faiss_index.reconstruct(8654)[None, :]
print(f"query embedding shape: {query_embedding.shape}")
query_embedding /= np.linalg.norm(query_embedding, axis=1, keepdims=True)
print(query_embedding)
```

```python

# get matches
similarities, indices = faiss_index.search(query_embedding, 3)  # type: ignore
print(f" indices:\n\t", indices[0])
print(f" similarities:\n\t", similarities[0])
for i in range(3):
    print(f" filenames:\n\t", filenames[indices[0][i]])

# get best match
best_match = os.path.join(
    input_path, basename(filenames[indices[0][0]]) + "_t00s00.mid"
)
best_similarity = similarities[0][0]

best_match, best_similarity
```
