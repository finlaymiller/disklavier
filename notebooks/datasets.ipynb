```python
import os
import shutil
from rich.progress import track
```

```python
def copy_every_third_file(source_dir, destination_dir):
    """
    Copies every third file from the source directory to the destination directory.

    :param source_dir: Path to the source directory
    :param destination_dir: Path to the destination directory
    """
    # Get a list of files in the source directory
    files = [
        f for f in os.listdir(source_dir) if os.path.isfile(os.path.join(source_dir, f))
    ]

    # Ensure the destination directory exists, create if not
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)

    # Copy every third file
    for i, file in track(enumerate(files)):
        if i % 3 == 0:
            shutil.copy(os.path.join(source_dir, file), destination_dir)


# Example usage
source_dir = "../inputs/all"
destination_dir = "../inputs/one third"
copy_every_third_file(source_dir, destination_dir)
```
